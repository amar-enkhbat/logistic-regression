## Introduction to Logistic Regression 

Logistic regression is NOT a regression model. Logistic regression is a binary classifier. For historical reasons, the Logistic Regression's regression term stuck. What makes logistic regression special from other classifiers such as perceptron and SVMs is that it outputs probabilistic results, which helps it to generalize more to non-linear problems. 

For example: Linear perceptron would classify a sample __x__ as either class 0 or 1. However, Logistic regression would result an output that looks like this: the probability that sample __x__ might belong to class 0 is 80% and class 1 is 20% etc.

### Why study Logistic Regression?

Logistic regression is important to learn because of following reasons:
  - It provides probabilistic idea to classification problem
  - It's a binary version of a Softmax classifier (which is used widely in deep learning algorithms)
  - Logistic regression's logistic function is also used in neural networks.

### The intuition and conditional probabilities of Logistic regression

Let's first assume that some data __x__ has a probability of P to belong to a class 0 (y = 0) when given weights __w__. We would write the conditional probability as,  $$P(y=0|\textbf{x};\textbf{w})$$ Then we introduce the _odds ratio_ which can be written as, $$\frac{P(y=0|\textbf{x};\textbf{w})}{1-P(y=0|\textbf{x};\textbf{w})}$$ Basically, what the odds ratio does is that it computes the quantity of odds in favor of particular event, the event in this case is the probability that __x__ might belong to class 0. 

We can then define the __logit__ function. Which is just the natural logarithm of the odds ratio. $$logit(P) = log \frac{P(y=0|\textbf{x};\textbf{w})}{1-P(y=0|\textbf{x};\textbf{w})}$$

```python
import numpy as np
import matplotlib.pyplot as plt

p = np.arange(0.01, 1, 0.01)
def odds_ratio(p):
    return p/(1-p)

def logit(p):
    return np.log(np.clip(odds_ratio(p), -250, 250))

plt.figure(figsize = (10, 4))

plt.subplot(121)
plt.plot(p, odds_ratio(p))
plt.xlabel("Probability")
plt.ylabel("Odds ratio")

plt.subplot(122)
plt.plot(p, logit(p))
plt.xlabel("Probability")
plt.ylabel("Logit")

plt.show()
```

As we can see from the above graph, the logit function maps probability domain to a nice domain where we can interpret the probabilities below 0.5 as negative numbers and above 0.5 as positive numbers.  

We can then invert the logit function to get the sigmoid function, $$P(y=0|\textbf{x};\textbf{w}) = \frac{1}{1+e^{logit(P)}}$$

Since the definition of a model hypothesis $z$ is $z = w^{T}x$ can take any real value we can substitute $logit(P)$ with our hypothesis $z$. Then we get the famous __Sigmoid function__, $$P(y=0|\textbf{x};\textbf{w}) = \frac{1}{1+e^{-z}}$$

```python
z = np.arange(-10, 10, 0.01)
def sigmoid(z):
    return 1/(1+np.exp(-z))
plt.plot(z, sigmoid(z))
plt.title("Logistic Function")
plt.xlabel("Hypothesis z")
plt.ylabel("Probability")
plt.show()
```

We talked about substituting a probability to different functions which might confuse many people. The gist is that, as we can see on the above graph, the Sigmoid function just maps the real values of hypothesis $z$ into a probabilistic domain, which we can then use to train the Logistic model.

## Logistic regression model

Now we will see how a logistic regression works. 

```python
from IPython.display import Image
Image(filename='images/logistic_regression_schematic.png', width=500)
```

Basically, what logistic regression does is that, <br> 
first, we compute the net sum $z$, <br>
then we feed $z$ through the activation function, which is the Sigmoid function, to get  our hypothesis $h$ or probabilistic results of each sample. <br>
Lastly, we compute the error and cost to feed it to the gradient descent algorithm to optimize the weights. <br>

### The hypothesis $h$

Given sample data $\textbf{x} = \{x^{(1)}, x^{(2)}, x^{(3)}, ..., x^{(m)}\}$ with $x^{(i)} \in \Re^{n}$ where $m$ is the number of samples and n is the number of features. <br>
and weights $\textbf{w} \in \Re^{(n \times 3)}$, <br>
the net sum is $z=\textbf{w}^T \textbf{x}=w_0x_0+w_1x_1+...+w_nx_n$ <br>
,where $x_0$ is the bias.
And our hypothesis $h=\phi(z)$ <br>
where, $\phi$ is the activation function aka the sigmoid function. <br>
Let's take dummy data and weigths to get a feel of how it works.

Assume we have 4 sample datas $\textbf{x}$ with two features and its corresponding class $\textbf{y}$ is as follows,

```python
x = np.array([[1, 1], [1, 2], [1, 4], [4, 1], [4, 3], [4, 4]])
y = np.array([0, 0, 0, 1, 1, 1])

plt.scatter(x[y==0, 0], x[y==0, 1], label = "class 0", marker = "x")
plt.scatter(x[y==1, 0], x[y==1, 1], label = "class 1", marker = "^")
plt.xlabel("feature 1")
plt.ylabel("feature 2")
plt.legend()
plt.show()
```

and the weights are,

```python
w = np.array([0.1, 0.2, 0.3])
print("Weights")
print(w)
```

then, the net sum $z$ is

```python
z = x.dot(w[1:]) + w[0]
print("Net sum of each sample is:")
print(z)
```

To compute our hypothesis we then feed our net sum z into the sigmoid function,

```python
h = sigmoid(z)
print("The hypothesis for each sample is:")
print(h)
```

We can then interpret the above output as, the probability that the samples $x^{(1)}, x^{(2)}, x^{(3)}, x^{(4)}$ belonging to class 1 is 65%, 71%, 86%, 89% and belonging to class 0 is 35%, 29%, 14%, 11% respectively.

We can then simplify this as, 
$$\widehat{y} = \left\{ \begin{array}{rcl}
{1} & \mbox{if}
& \phi(z) \geq 0.5 \\ 0 & \mbox{otherwise} &
\end{array}\right.
$$
if we look at the plot of the sigmoid function this is equivalent to,
$$\widehat{y} = \left\{ \begin{array}{rcl}
{1} & \mbox{if}
& z \geq 0 \\ 0 & \mbox{otherwise} &
\end{array}\right.
$$
where $\widehat{y}$ is the predicted classes.

If we apply it to the dummy data we get the predicted class labels:

```python
print("y_hat = ", (z >= 0) * 1)
```

we get that given weights $w=\{0.1, 0.2, 0.3\}$ we predict that all the dummy sample data belong to class 1, which is wrong. <br>
So we have to tune the weight so that $\widehat{y} = \{0, 0, 1, 1\}$ and comes the next part. The training.

### Training the weights

Now we can calculate the probabilities and predict the class labels. However, predicted labels were wrong. We now have to train our weights so that it can correctly classify class labels.

#### The cost function

The cost function of the logistic regression is called the __log-likelyhood function__. The formula is a follows:
$$J(\textbf{w})=-\sum_i ( y^{(i)} log(\phi (z^{(i)}))+(1-y^{(i)})log(1-\phi (z^{(i)})))$$
We can see that the first term becomes 0 if $y=0$, and the second term becomes zero if $y=1$:
$$J(\phi(z), y;\textbf{w}) = \left\{ \begin{array}{rcl}
{-log(\phi(z))} & \mbox{if} & y=1 \\ 
-log(1-\phi(z)) & \mbox{if} & y=0
\end{array}\right.
$$
Basically, what it does is that when we get a wrong prediction the cost increases.
Let's calculate the cost of our dummy data to see how much our prediction is wrong.

```python
def compute_cost(z):
    return np.sum(-y*(np.log(sigmoid(z))) - ((1 - y)*(np.log(1 - sigmoid(z)))))
print("Total cost:", compute_cost(z))
```

Our cost was equal to 2.55. We can then use this to update our weights so that the hypothesis give as less cost as possible.

#### Updating the weights

In our logistic regression model, we will use gradient descent algorithm to update our weight so that our models hypothesis gives as low cost as possible.
The algorithm for gradient descent is as follows:
$$\textbf{w} = \textbf{w} -\eta \Delta J(\textbf{w})$$
where, $\eta$ is the learning rate and $\Delta J(\textbf{w})$ is the derivative of our cost function. 

The idea behind gradient descent is that by calculating the _gradient_ or the _slope_ of the cost function we can get a general idea of which way direction is optimal for reducing the cost. We can imagine the cost function as a valley the weights as a position $\Delta J(\textbf{w})$ as the direction and $\eta$ as the size of our step. <br>
Imagine you're blindfolded and placed on top of a hill and you have to go down without looking. By taking a small step you can get the general idea which way is down by the slope of the hill.

```python
Image(filename='images/gradient_descent.png', width=500)
```

The derivative of the cost function is as follows,
$$\Delta J_j({\textbf{w}}) = -\sum_i (y^{(i)} - \phi(z^{(i)}))x^{(i)}_j$$

#### Training epochs and convergence

Now we can a logistic regression model however, there's a tiny problem. We have the algorithm to take a small step down the hill however we don't have any means to stop. We can define the amount of steps we take or iterations or _epochs_ we can stop at the general local minima of our cost function. Also, we can give our model a condition to stop such as: if the difference between the cost in last two iterations is smaller than $\epsilon$ (for example, $\epsilon < 10^{-4}$)then stop training etc. 

#### The training

Now let's train a logistic regression model using our dummy samples.

```python
# Number of iterations
epoch = 100

# Learning rate
learning_rate = 0.001

# Cost array
cost_array = []

# Initialize the weights
w = np.array([0, 0, 0], dtype = float)

# Training
for _ in range(epoch):
    # Dot product of z (= w1*x1 + w2*x2 + ... + wn*xn) + w0
    z = np.dot(x, w[1:]) + w[0]
    
    # Cost function (Cross Entropy Function)
    cost = compute_cost(z)
    cost_array.append(cost)
    
    # Weight update 
    w[0] -= learning_rate * np.sum(sigmoid(z) - y)
    w[1:] -= learning_rate * (np.dot((sigmoid(z) - y), x))
        
    # Convergence condition
    if(_ > 2):
        if (cost_array[-2] - cost_array[-1]) < 0.001:
            break
learned_weights = w
print("Learned weights: ", learned_weights)
```
